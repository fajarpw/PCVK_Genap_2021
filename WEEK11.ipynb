```python
# Import Library
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Load Image
cats_and_bunnies = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/cats_and_bunnies.jpg')
parking_lot = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/parking-lot-cars.jpg')
chess_board = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/chess-board.jpg')
sudoku= cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/sudoku.jpg')
laptop = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/laptop.jpg')
```

```python
# Input Image and Template Image
cats_and_bunnies = cv.cvtColor(cats_and_bunnies, cv.COLOR_BGR2RGB)

cat_template = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/cat2_templatejpg.jpg', 0)

plt.subplot(1,2,1),plt.imshow(cats_and_bunnies)
plt.title('Input Image')
plt.subplot(1,2,2),plt.imshow(cat_template, cmap='gray')
plt.title('Template Iamge')

plt.show()
```

```python
# cv.TM_CCOEFF
cat_1 = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/cats_and_bunnies.jpg',0)
w, h = cat_template.shape[::-1]

result_cat = cv.matchTemplate(cat_1, cat_template, cv.TM_CCOEFF)

min_val, max_val, min_loc, max_loc = cv.minMaxLoc(result_cat)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv.rectangle(cats_and_bunnies, top_left, bottom_right, 255, 5)

plt.subplot(1,2,1),plt.imshow(result_cat, cmap='gray')
plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(cats_and_bunnies)
plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
plt.suptitle('cv.TM_CCOEFF')

plt.show()
```

```python
# Sobel Edge Detection
gray_parking_lot = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/parking-lot-cars.jpg',0)
edges = cv.Sobel(gray_parking_lot, cv.CV_8U, 1, 0, ksize=5)

plt.subplot(121),plt.imshow(parking_lot)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Canny Edge Detection
gray_parking_lot = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/parking-lot-cars.jpg',0)
edges = cv.Canny(gray_parking_lot,100,200)

plt.subplot(121),plt.imshow(parking_lot)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Harris Corner Detection
gray_chess_board = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/chess-board.jpg',0)

gray_chess_board = np.float32(gray_chess_board)
dst = cv.cornerHarris(gray_chess_board,2,3,0.04)

dst = cv.dilate(dst,None)

chess_board[dst>0.01*dst.max()]=[0,0,255]
cv2_imshow(chess_board)
```

```python
# Shi-Tomasi Edge Detection
gray_chess_board = cv.imread('/content/drive/MyDrive/PCVK/Images/Object Detection/chess-board.jpg',0)
chess_board = cv.cvtColor(chess_board, cv.COLOR_BGR2RGB)
corners = cv.goodFeaturesToTrack(gray_chess_board,25,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv.circle(chess_board,(x,y),3,255,-1)

cv2_imshow(chess_board)
```
