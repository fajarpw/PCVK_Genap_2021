```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load image
img = cv.imread('/content/drive/MyDrive/PCVK/Images/j.png')
cv2_imshow(img)
```

```python
# 5 - Dilasi SE Square 5x5 
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation')
plt.show()
```

```python
# 5 - Erotion SE Square 5x5 
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(erosion),plt.title('Erotion')
plt.show()
```

```python
# 7 - Opening SE Square 7x7
kernel = np.ones((7,7),np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(opening),plt.title('Opening')
plt.show()
```

```python
# 8 - Closing SE Square 7x7
kernel = np.ones((7,7),np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(opening),plt.title('Opening')
plt.show()
```

```python
# 9 Dilation Cross 3x3 
kernel = cv.getStructuringElement(cv.MORPH_CROSS,(3,3))
dilation = cv2.dilate(img,kernel,iterations = 1)
 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation')
plt.show()
```
